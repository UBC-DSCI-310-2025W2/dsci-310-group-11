```python
from pathlib import Path
import pandas as pd

# make sure data directory exists
Path("data").mkdir(parents=True, exist_ok=True)

# fetch dataset from UCI
wine_quality = fetch_ucirepo(id=186)

# features and target
X = wine_quality.data.features
y = wine_quality.data.targets

# combine them into one dataframe
df = pd.concat([X, y], axis=1)

# inspect metadata and variables
print(wine_quality.metadata)
print(wine_quality.variables)

print(df.head())
print(df.columns)
print(df.shape)
```

```python
from pathlib import Path

# if a color/type column exists, keep only red wine
for possible_col in ["color", "type", "wine_type"]:
    if possible_col in df.columns:
        df = df[df[possible_col].astype(str).str.lower() == "red"].copy()
        df = df.drop(columns=[possible_col])
        break

output_path = Path("data/winequality-red.csv")
df.to_csv(output_path, index=False)

print(f"Saved dataset to {output_path}")
print(df.head())
print(df.shape)
```
